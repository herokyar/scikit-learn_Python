```python
#create input data matrix and the output data
titanic_X, titanic_y = df[["pclass","age","sex"]], df['survived']
```

```python
#encode sex into binary representation

enc = LabelEncoder()
label_encoder = enc.fit(titanic_X['sex'])
print 'categorical classes', label_encoder.classes_
integer_classes = label_encoder.transform(label_encoder.classes_)
print 'integer classes:', integer_classes
t = label_encoder.transform(titanic_X['sex'])
titanic_X['sex'] = t
```

```python
#sex has been encoded from string into binary numbers [0,1]
titanic_X.head()
```

```python
#Label encoding with DictVectorizer !!!

#turn training set into list of dictionaries
titanic_X.transpose().to_dict().values()
```

```python
#get the encoded input matrix
dvec = DictVectorizer(sparse=False)
X = dvec.fit_transform(titanic_X.transpose().to_dict().values())
print X
```

```python
#convert X into pandas dataframe again
titanic_X = pd.DataFrame(X, columns=dvec.get_feature_names())
```

```python
#plass column is encoded.
titanic_X.head(20)
```

```python
#some more information about new dataframe
titanic_X.describe()
```

```python
#check if any cell in the age column has NAN values
titanic_X['age'].isnull().values.sum()
```

```python
#fill NAN values with the median age value
print titanic_X['age'].mean()

titanic_X['age'] = titanic_X['age'].fillna(titanic_X['age'].mean())
```

```python
#see that all the NAN values are replaced with the age mean value #for example check age index 12, 13, 14 
titanic_X.head(20)
```

```python
#split the dataset into train and test set
X_train, X_test, y_train, y_test = train_test_split(titanic_X, titanic_y, test_size=0.25, random_state=33)
```

```python
#create a new tree classifier
clf = tree.DecisionTreeClassifier(criterion='entropy', max_depth=3, min_samples_leaf=5)
```

```python
#fit the model
clf = clf.fit(X_train, y_train)
```

```python
#returns an error -> global name 'dot_parser' is not defined
dot_data = StringIO.StringIO() 
tree.export_graphviz(clf, out_file=dot_data, feature_names=['age','sex','1st_class','2nd_class','3rd_class']) 
graph = pydot.graph_from_dot_data(dot_data.getvalue()) 
graph.write_png('titanic.png') 
from IPython.core.display import Image 
Image(filename='titanic.png')
```

```python
#function for measuring accuracy on the training set

def measure_performance(X,y,clf, show_accuracy=True, show_classification_report=True, show_confusion_matrix=True):
    y_pred=clf.predict(X)   
    if show_accuracy:
        print "Accuracy:{0:.3f}".format(metrics.accuracy_score(y,y_pred)),"\n"

    if show_classification_report:
        print "Classification report"
        print metrics.classification_report(y,y_pred),"\n"
        
    if show_confusion_matrix:
        print "Confusion matrix"
        print metrics.confusion_matrix(y,y_pred),"\n"
 
```

```python
#get the accuracy     
measure_performance(X_train,y_train,clf, show_classification_report=False, show_confusion_matrix=False)
```

```python
#apply cross validation on the training set
accuracies = cross_val_score(clf, X_train, y_train,cv=5)
```

```python
print 'Cross validation accuracy:', np.mean(accuracies), 
accuracies
```
