```python
#Banner Blocker on web pages
#using python scikit-learn and decision trees, random forests as classifiers
#(which are   non-linear classifiers)

#download the database at: http://archive.ics.uci.edu/ml/datasets/Internet+Advertisements

#this model will classify the adds shown on web pages as 'ad' or 'no-ad' 

#to learn more about the structure of the dataset, refer to the readme file 
#here is the description: http://archive.ics.uci.edu/ml/machine-learning-databases/internet_ads/ad.DOCUMENTATION

#from the database description:

#5. Number of Instances: 3279 (2821 nonads, 458 ads)
#6. Number of Attributes: 1558 (3 continous; others binary; this is the
#   "STANDARD encoding" mentioned in the [Kushmerick, 99].)
#   One or more of the three continous features are missing in 28%
#   of the instances; missing values should be interpreted as "unknown".


 
```

```python
#import some libraries
import pandas as pd

from sklearn.tree import DecisionTreeClassifier
from sklearn.cross_validation import train_test_split
from sklearn.metrics import classification_report
from sklearn.pipeline import Pipeline
from sklearn.grid_search import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

from IPython.display import Image
```

```python
#get the data into Pandas dataframe,  see that the data does not have any headers
df = pd.read_csv('ad.data', header=None)
```

```python
df.describe() #this is a 10x1554 data
```

```python
df.head(10) #get the first 10 examples in the data
```

```python
print 'there are', len(df), 'examples in the dataset.'
```

```python
#get the features 
explanatory_variable_columns = set(df.columns.values)
```

```python
#set the output column of the dataframe
response_variable_column = df[len(df.columns.values)-1]
```

```python
#the last column describes the targets!! 
#so remove it from explanatory_variable_columns
explanatory_variable_columns.remove(len(df.columns)-1)
```

```python
#create the output labels (turn the non-numeric values to numeric values so that scikit-learn can read)
# if ad -> 1
# else -> 0
y = [1 if e=='ad.' else 0  for e in response_variable_column] #list comprehension
```

```python
#create the features of sub data-frame 
X= df[list(explanatory_variable_columns)]
```

```python
#there are many missing values in the dataframe 
# these values can be replaced by -1
#or simply we can assign the mean value of the column for each missing cell
X.replace(to_replace= ' *\?', value=-1, regex=True, inplace=True)
```

```python
X.head(11) #you can see that the '?' values in the dataframe are replaced with -1 value (for ex: check out row number 10)
```

```python
#split the data into training and test sets

X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```python
#create a pipeline
pipeline = Pipeline([

('clf', DecisionTreeClassifier(criterion='entropy')) #builds the decision tree using the information gain heuristic

])
```

```python
#specify the hyper parameters for the model
parameters = {
'clf__max_depth': (150, 155, 160),
'clf__min_samples_split': (1, 2, 3),
'clf__min_samples_leaf': (1, 2, 3)
}
```

```python
#train your model using decision tree and scikit learn's Grid Search
grid_search = GridSearchCV(pipeline, parameters, n_jobs=-1, verbose=1, scoring='f1')
grid_search.fit(X_train, y_train)
print 'best score: %0.3f' % grid_search.best_score_
print 'best parameters set:'
best_parameters = grid_search.best_estimator_.get_params()

for param_name in sorted(parameters.keys()):
    print '\t%s: %r' % (param_name, best_parameters[param_name])
    

predictions = grid_search.predict(X_test)
print classification_report(y_test, predictions)
```

```python
#you have to actually run this line of code outside of ipython notebook in the command window to get the best parameters(parallel computing)

import pandas as pd

from sklearn.tree import DecisionTreeClassifier
from sklearn.cross_validation import train_test_split
from sklearn.metrics import classification_report
from sklearn.pipeline import Pipeline
from sklearn.grid_search import GridSearchCV


if __name__ == '__main__':

    df = pd.read_csv('ad.data', header=None)

    explanatory_variable_columns = set(df.columns.values)
    response_variable_column = df[len(df.columns.values)-1]

    explanatory_variable_columns.remove(len(df.columns)-1)

    y = [1 if e=='ad.' else 0  for e in response_variable_column]

    X= df[list(explanatory_variable_columns)]

    X.replace(to_replace= ' *\?', value=-1, regex=True, inplace=True)

    X_train, X_test, y_train, y_test = train_test_split(X,y)

    pipeline = Pipeline([

    ('clf', DecisionTreeClassifier(criterion='entropy'))
    ])


    parameters = {
    'clf__max_depth': (150, 155, 160),
    'clf__min_samples_split': (1, 2, 3),
    'clf__min_samples_leaf': (1, 2, 3)
    }

    grid_search = GridSearchCV(pipeline, parameters, n_jobs=-1, verbose=1, scoring='f1')
    grid_search.fit(X_train, y_train)
    print 'best score: %0.3f' % grid_search.best_score_
    print 'best parameters set:'
    best_parameters = grid_search.best_estimator_.get_params()

    for param_name in sorted(parameters.keys()):
        print '\t%s: %r' % (param_name, best_parameters[param_name])
        

    predictions = grid_search.predict(X_test)
    print classification_report(y_test, predictions)



```

```python
#here is the line of code with Random Forests  (run this line of code in the console)

import pandas as pd

from sklearn.tree import DecisionTreeClassifier
from sklearn.cross_validation import train_test_split
from sklearn.metrics import classification_report
from sklearn.pipeline import Pipeline
from sklearn.grid_search import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

if __name__ == '__main__':

    df = pd.read_csv('ad.data', header=None)

    explanatory_variable_columns = set(df.columns.values)
    response_variable_column = df[len(df.columns.values)-1]

    explanatory_variable_columns.remove(len(df.columns)-1)

    y = [1 if e=='ad.' else 0  for e in response_variable_column]

    X= df[list(explanatory_variable_columns)]

    X.replace(to_replace= ' *\?', value=-1, regex=True, inplace=True)

    X_train, X_test, y_train, y_test = train_test_split(X,y)

    pipeline = Pipeline([

    ('clf', RandomForestClassifier(criterion='entropy'))
    ])


    parameters = {
    'clf__n_estimators': (5, 10, 20, 50),
    'clf__max_depth': (50, 150, 250),
    'clf__min_samples_split': (1, 2, 3),
    'clf__min_samples_leaf': (1, 2, 3)
    }

    grid_search = GridSearchCV(pipeline, parameters, n_jobs=-1, verbose=1, scoring='f1')
    grid_search.fit(X_train, y_train)
    print 'best score: %0.3f' % grid_search.best_score_
    print 'best parameters set:'
    best_parameters = grid_search.best_estimator_.get_params()

    for param_name in sorted(parameters.keys()):
        print '\t%s: %r' % (param_name, best_parameters[param_name])
        

    predictions = grid_search.predict(X_test)
    print classification_report(y_test, predictions)



```
