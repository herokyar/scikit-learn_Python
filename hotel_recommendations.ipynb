# #recommendation algorithm

#### here a different approach from usual collaborative filtering algorihm has been followed. because each user does not have different ratings for each hotel (hotel ratings are fixed), some similarity metrics such as Euclidian or Pearson correlation could not be applied on the data to make a recommendation.  the idea here is that, for a unique user, all the other users who liked same hotels have been investigated. for example if one user liked hotel numbers 1,2 and 3, all the other users who also liked hotel number 1,2,3 have been found. there is a sorting criteria here that, the fist 10 most similar users have been selected. that means , those 10 users have the highest commonality with the selected user. after getting those most similar users, also most common liked hotels between those 10 users have been searched. as a prediction of the hotel for that particular user that he/she will look at next, the first hotel from the most liked hotels between those 10 users is taken as the predicted value.   as an improvement, istead of taking the most similar 10 users, that number can be increased, for example most similar 20 users to that particular user can be taken also gender information of the users could have been used in order to make a better prediction.

```python
def getRecommendation(dictpersons, person):
    
    dicta = {}  #intersection between users that liked the same hotel
    
    for i in range(len(dictpersons)):
    
        a = dictpersons[i+1].keys()
        b = dictpersons[person].keys()

        c = (set(a) & set(b))

        if len(c) > 0:
            dicta[i+1] = list(c) 
        
    
    top_10_similar_users = []

    for k in sorted(dicta, key=lambda k: len(dicta[k]), reverse=True):
        top_10_similar_users.append([k, dicta[k]])
    
    top_10_similar_users = top_10_similar_users[:10] #this number might change
    
    
    bdict = {}

    for u in top_10_similar_users: #get the total likes for each hotel 

        for hotelid in  dictpersons[u[0]].keys():
            if hotelid not in bdict:
                bdict[hotelid] = 1
            else:
                bdict[hotelid] += 1

    #print bdict
    
    
    blist = []

    for w in sorted(bdict, key=bdict.get, reverse=True):
        if w in newdict[person].keys(): #dont take the hotels user already liked
            continue
        else:
            #print w, bdict[w]
            blist.append([w, bdict[w]])

            
    return blist[0][0] #take the first element, recommend this hotel to the user
```

# #find predictions for every user

```python
predictions_dict = {}

for u in range(len(newdict)):
        
    prediction_value = getRecommendation(newdict, u+1)
    
    predictions_dict[u+1] = prediction_value
    
        
```
