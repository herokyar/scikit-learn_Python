```python
#image quantization using k-means unsupervised clustering algorithm (K=64)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.utils import shuffle
import mahotas as mh

import os
```

```python
#create a function for image quantization

def imageQuantize(image_file, no_of_clusters=64): #default cluster size is 64
    
    #read the original image into numpy array
    original_img = np.array(mh.imread(image_file), dtype=np.float64) / 255
    
    #get the original dimensions in tuple format
    original_dimensions = tuple(original_img.shape)
    
    #extract width, height and depth of the image
    width, height, depth = tuple(original_img.shape)
    
    #flatten the image array
    image_flattened = np.reshape(original_img, (width * height, depth))
    
    #get a subset (1000 colors) from the image (in random order)
    image_array_sample = shuffle(image_flattened, random_state=0)[:1000]
    
    #create a k-means estimator with K=64 clusters
    #changing K value will change the quality of the compression
    estimator = KMeans(n_clusters=no_of_clusters, random_state=0) 
    
    #train the model
    estimator.fit(image_array_sample)
    
    #assign each color in the original image a cluster
    cluster_assignments = estimator.predict(image_flattened)
    
    #get the cluster centers (centroids)
    compressed_palette = estimator.cluster_centers_
    
    #initialize compressed image to zero numpy array
    compressed_img = np.zeros((width, height, compressed_palette.shape[1]))
    
    #initialize label to value 0
    label_idx = 0
    
    #recreate the compressed image
    for i in range(width):
        for j in range(height):
            compressed_img[i][j] = compressed_palette[cluster_assignments[label_idx]]
            label_idx += 1
            
    #plot the original and compressed images together
    plt.subplot(121)
    plt.title('original image')
    plt.imshow(original_img)
    plt.axis('off')
    plt.subplot(122)
    plt.title('compressed image')
    plt.imshow(compressed_img)
    plt.axis('off')
    plt.show()   
    
```

```python
#get the file list in the images folder 
file_list = os.listdir("C:/Users/hasan/images/")
```

```python
#itarete over the images folder
for file in file_list:
    imageQuantize("C:/Users/hasan/images/"+file, 32) #use 32 cluster for compression
```
