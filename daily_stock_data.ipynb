# #web scraping, plotting and writing into a txt file daily stock data for companies

```python
import numpy as np
import pandas as pd

from datetime import datetime

import urllib
import re
import json
```

```python
company_names = ['AAPL', 'GOOG', 'SPY','NFLX'] #add any company you want into the list
```

```python
def plot_daily_stock(company):
    
    #added later, write data into a text file
    myfile = open('%s.txt' % company, 'w+') #w+ means create file and write 
    myfile.close()
    
    htmltext = urllib.urlopen('http://www.bloomberg.com/markets/chart/data/1D/%s:US' % company)
            
    data = json.load(htmltext)    

    datapoints =  data['data_values']
    
    #added later
    myfile = open('%s.txt' % company, 'a') #a means append    
        
    company_daily_data = []  #create an empty list
    
    
    for point in datapoints:
    #print point[1]
        company_daily_data.append(point[1])
        myfile.write(company + ',' + str(point[0]) + ',' + str(point[1]) + '\n') #this way creates csv file, important add '\n' at the end of the file
    myfile.close() #after writing everything, close the file
        
    #stock_index = pd.date_range('5/27/2015 9:30', '5/27/2015 15:59', freq='1min') #you can also automatically update the time
    stock_index = pd.date_range('5/27/2015 9:30', periods=len(datapoints) , freq='1min') #use this version because period might change
    
    df_example = pd.DataFrame(company_daily_data, index=stock_index, columns=[company]) #or instead of dataframe you can use series
    
    df_example.plot()
    plt.title('%s daily data' % company)
```
